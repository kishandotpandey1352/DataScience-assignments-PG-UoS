```python
def get_yearly_EU_df(year):
    """
    This function takes the year value as input and return a dataframe with that year data rows, from European nations and
    adds a new coulmn 'co2_percentage'
    
    Parameters
        ----------
        year : integer
            The year for which the filtering of data needed
    Returns
        ----------
        df_temp :dataframe
            A temporary dataframe of selected year andEuropean nations
    """
    df_temp = df.loc[((df['country'].isin(EU_nations)) & (df['year']==year))].copy()
    df_temp.dropna(subset=['co2'], inplace=True)
    year_specific_year_EU_co2 = df.loc[(df['country']=='Europe') & (df['year']==year)]['co2'].values[0]
    # year_specific_year_EU_co2
    df_temp["co2_percentage"] = df_temp['co2']/year_specific_year_EU_co2*100
    return df_temp
```

```python
def plot_bar_graph(x,y,plot_title,ax,bgc,barcolor):
    """
    A function to plot the bar grapgh when the data for x & Y co-ordinates provided 
    Parameters
        ----------
        x : A pandas series
            data for x axis    
        y: A pandas series
            data for y axis  
        plot_title: 
        ax: Axes
            An axis of the matplot figure
        bgc: String
            A color to set background color of the chart
        barcolor: a list
            A list of colors for bars
    Returns:None
        ----------
        
    """
    color = barcolor
    #set face color of the canvas
    ax.set_facecolor(bgc)
    ax.set_yticks((np.arange(0, 100, 10))) 
    #plot bar graph
    ax.bar(x, y, label=x,color=color,width=0.5,align='center')
    # set y axis label
    ax.set_ylabel('Percentage of world CO2 emissions')
#     ax.set_xlabel('Top Five countries with higher CO2 emissions')
    #set title
    ax.set_title(plot_title)
    ax.grid(axis='y')

#     ax.legend(title="Countries")
    

def get_yearly_df(year, type):
    """
    This function takes the year value as input and return a dataframe with that year data rows and adds a new 
        coulmn co2_percentage
    """
    df_temp = df.loc[df['year']==year].copy()
    year_specific_world_co2 = df_temp.loc[df_temp['country']==type,['co2']].values[0]
    df_temp["co2_percentage"] = df_temp['co2']/year_specific_world_co2*100
    return df_temp
```

## Q3. The link between GDP and CO<sub>2</sub> emissions

*Do countries with high GDP always have high per capita CO<sub>2</sub> emissions.* 

Plot the per capita CO<sub>2</sub> emissions against the GDP per capita for each country, adjusting the area of each marker to represent the population of the country. Only include countries with a population of at leat 10 million people in 2018. Annotate significant countries on the plot, such as large countries or outliers.

Compare these plots for the year 1978, 1998 and 2018. Comment on any patterns in the relationship between GDP per capita and CO emissions per capita and how they may have changed over time.

(Add your solution in the cells provided below.)

```python
# ADD YOUR CODE HERE. THE CODE SHOULD PRODUCE A SINGLE FIGURE.
# 2018
population = 10**7
def gdp_co2_yearwise_data(year):
    '''
    A function that takes an year as an argument and returns a dataframe with new coulmns 
    (columns: gdp_per_capita,co2_per_capita) created 
    
    Parameters
        ----------
        year : integer
            The year for which the data needs to be extracted
    Returns
        ----------
        df_temp :dataframe
            The filtered dataframe with additional rows created
    '''
    df_temp = df.loc[(df['year']==year) & (df['population']>=population)].copy()
    # delete the rows for continents, as we are focussed on individiual countries
    df_temp.dropna(subset=['iso_code'], inplace=True) 
    # add 2 coulmns 'gdp_per_capita' & 'co2_per_capita' for further analysis
    df_temp['gdp_per_capita'], df_temp['co2_per_capita'] = df_temp['gdp']/df['population'], df_temp['co2']/df['population']
    return df_temp
```

```python
#significant countries calculations
def significant_countries_by_year(year):
    '''
    A function that takes an year as an argument and returns a set of significant country names
    
    Parameters
        ----------
        year : integer
            The year for which the important country names needed
    Returns
        ----------
        df_temp :set
            Set of significant countries to assist in annotation
    '''
    # extract data for the specified year
    df_temp = gdp_co2_yearwise_data(year)
    #gdp wise 4 significant countries
    df_temp = df_temp.sort_values(by='gdp_per_capita', key=abs, ascending = False)
    significant_countries_gdp_wise = pd.Series(df_temp['country'][:4])#top 5 with high gdp per capita
    #co2 wise 4 significant countries
    df_temp = df_temp.sort_values(by='co2_per_capita', key=abs, ascending = False)#top 5 with high CO2 per capita
    significant_countries_co2_wise = df_temp['country'][:4]
    #Population wise 4 significant countries
    df_temp = df_temp.sort_values(by='population', key=abs, ascending = False)#top 5 with high CO2 per capita
    significant_countries_population_wise = df_temp['country'][:4]
    # concatenate all significant countries to get a set
    significant_countries = pd.concat([significant_countries_gdp_wise,significant_countries_co2_wise,significant_countries_population_wise])
    return set(significant_countries)
```

```python
def plot_bubble_diagram(data,significant_countries,year):
    """
    A function to plot bubble/scatter plot for co2_per_capita vs gdp_per_capita data """
    #data
    x = data['gdp_per_capita']
    y = data['co2_per_capita']
    selected_index = data.loc[data['country'].isin(significant_countries)].index
    #scatter plot - bubble blot
    plt.scatter(x, y,s=data['population']*10**(-7),
       alpha=0.4,
       linewidth=2)
    # annotation
    for i in list(selected_index):
        plt.annotate(f"{data.loc[i]['country']}", (x[i], y[i]), xytext=(0, 2), ha='center',
                     annotation_clip = True,textcoords='offset fontsize',
                     arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))

    # Add title and labels
    plt.xlabel("gdp per capita")
    plt.ylabel("co2 per capita")
    plt.title(f"{year} ::: co2 per capita vs gdp per capita over country population")

    # Show the graph
    plt.show()
```

```python
### Plot the diagram for each year ###
significant_countries = significant_countries_by_year(2018)
# data # countries of World in 1978
fig, axes = plt.subplots(figsize=(15, 8), sharey = True,facecolor='lightpink')
year = 1978
data_1978 = gdp_co2_yearwise_data(year)
# significant_countries = significant_countries_by_year(2018)
plot_bubble_diagram(data_1978,significant_countries, year)
# data # countries of World in 1998
fig, axes = plt.subplots(figsize=(15, 8), sharey = True,facecolor='lightgreen')
year = 1998
# significant_countries = significant_countries_by_year(year)
data_1998 = gdp_co2_yearwise_data(year)
plot_bubble_diagram(data_1998,significant_countries, year)
# data # countries of World in 2018
fig, axes = plt.subplots(figsize=(15, 8), sharey = True,facecolor='skyblue')
year = 2018
data_2018 = gdp_co2_yearwise_data(year)
plot_bubble_diagram(data_2018,significant_countries,year)
```

## Q4. The development of wealth inequality over time.

*How has wealth and wealth inequality developed over time?*

Make a plot that shows the distribution of GDP per capita across countries at each decade from 1950 to 2020 (i.e., at the years 1950, 1960, 1970, etc.). Only include countries with a population of at least 5 million in 2020.

The World Bank defines an income of less than $2.15 per day as extreme poverty, corresponding to an annual GDP per capita of $785. Mark this threshold on your plots. For each year shown, calculate the number of people living in countries with a GDP per capita below this level and annotate this in your figure. Comment on whether this population has grown or shrunk as a proportion of the worlds total population over time.

(Add your solution in the cells provided below.)

```python
# ADD YOUR CODE HERE. THE CODE SHOULD PRODUCE A SINGLE FIGURE.
# Filter the data based on the year (2020) and where population is at least five million
df_4 = df.loc[(df['year']==2020) & (df['population']>=5*10**6)].copy()
df_4.dropna(subset=['iso_code'], inplace=True) # removed empty rows
# df_4.head(7)
```

```python
# Consider the countries with a population of at least 5 million in 2020
selected_countries_2020 = list(df_4['country'].unique())
```

```python
# Filter data in the year range of 1950 to 2020 and from the selected countries from year 2020 (selected_countries_2020)
df_4a = df.loc[((df['year']>=1950) & (df['year']<=2020) & (df['population']>=5*10**6) & (df['country'].isin(selected_countries_2020)))].copy()
# Remove empty rows to take just countries
df_4a.dropna(subset=['iso_code'], inplace=True) 
# add a new column to store annual gdp per capita
df_4a['annual_gdp_per_capita'] = df_4a['gdp']/df_4a['population']
# Convert the 'year' column to decade format
df_4a['decade'] = (df_4a['year'] // 10) * 10
# df_4a
# df_4['country'].value_counts()
```

```python
#number of people living in these countries whose annual_gdp_per_capita is less than 785
poor_country_group = df_4a[df_4a['annual_gdp_per_capita'] < 785].groupby(['decade'])['population'].sum().reset_index(name='poor_people_count')
# poor_country_group
```

```python
# Use seaborn to create a boxplot of GDP per capita across decades
# sns.boxplot(data=df_4a, x='decade', y='annual_gdp_per_capita')
fig, ax = plt.subplots(figsize=(10, 15),facecolor='lightpink')
ax = sns.boxplot(data=df_4a, x='decade', y='annual_gdp_per_capita',palette="coolwarm", hue ="decade",legend = True)
# draw a line at threshold annual_gdp_per_capita of $785
# sns.lineplot(data = df_4a,x = 'decade', y = 785)
# Set plot labels and title
ax.axhline(785, color="r", dashes=(2, 2))
poor_country_group['poor_people_count'].plot(x=ax, kind='line', marker='o', color='green')
#annotate the number of countries in each decade with annual_gdp_per_capita < $785
ax.annotate('$ 785 - GDP threshold (line)', xy=(7, 785), xytext=(10, 500),
            arrowprops=dict(facecolor='red', shrink=5))
#annotate the points representing number of people falling in poor category in each decade
for i, txt in enumerate(poor_country_group['poor_people_count']):
    plt.annotate(
        f'{poor_country_group['poor_people_count'][i]}', 
        (poor_country_group['poor_people_count'].index[i],
        poor_country_group['poor_people_count'][i]),
        textcoords="offset points", 
        xytext=(-10, -20), 
        ha='center')


plt.title('Distribution of GDP per Capita Across Countries by Decade')
plt.xlabel('Decade')
plt.ylabel('GDP per Capita')
#show legend
lines = [] 
labels = []

for ax in fig.axes: 
    Line, Label = ax.get_legend_handles_labels() 
    # print(Label) 
    lines.extend(Line) 
    labels.extend(Label) 
  
# rotating x-axis labels by 45 degree
plt.xticks(rotation=45) 
# show verticle grid lines
plt.grid(axis="x")
# Show the plot
plt.xticks(rotation=0)
plt.legend(loc=0)
# Use a logarithmic scale for better visualization of wide range on y-axis
plt.yscale('log')  
plt.show()
```
