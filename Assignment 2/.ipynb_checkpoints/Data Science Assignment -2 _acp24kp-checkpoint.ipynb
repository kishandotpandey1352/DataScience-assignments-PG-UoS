```python
# A function to generate augmented images
def generate_augmented_images(X):
    '''
    A function to generate augmented images
    '''
    # Set the random seed for reproducibility
    np.random.seed(42) 
    
    # Copy the original array
    x_train_new = X.copy()

    for i, image in enumerate(X):
        # Initialize an empty array for the augmented image
        augmented_image = np.empty((5829, ))  
        
        # Split image pairs into two parts for two images
        image_1 = image[:2914].reshape(62, 47)
        image_2 = image[2914:].reshape(62, 47)

        # Convert the numpy arrays to Image objects for transformations
        image1 = Image.fromarray(image_1)
        image2 = Image.fromarray(image_2)
        
        # Horizontal flip
        if random.choice([True, False]):
            image1 = image1.transpose(Image.FLIP_LEFT_RIGHT)
            image2 = image2.transpose(Image.FLIP_LEFT_RIGHT)

        # Vertical flip 
        if random.choice([True, False]):
            image1 = image1.transpose(Image.FLIP_TOP_BOTTOM)
            image2 = image2.transpose(Image.FLIP_TOP_BOTTOM)

        # Rotate images
        angle = random.choice([0, 90, 180, 270])
        image1 = image1.rotate(angle)
        image2 = image2.rotate(angle)

        # Convert back to numpy array after transformation
        image1 = (np.array(image1)).reshape(1, -1)
        image2 = (np.array(image2)).reshape(1, -1)
        
        # Combine images into one augmented pair
        augmented_image = np.hstack((image1, image2))
        # Append the augmented image to the new dataset
        x_train_new = np.vstack((x_train_new, augmented_image))

    return x_train_new
```

```python
# A function to augment the labels, similar to image pairs array augmentation
def generate_augmented_labels(Y):
    y_train_new = Y.copy()
    augmented_labels = Y
    y_train_new = np.vstack((y_train_new, augmented_labels))
    return y_train_new    
```

```python
# Update the FunctionTransformer to work with both features and labels
def generate_augmented_data(X, n_times = 2):
    # Global Y to keep single variable to store changes labels
    global Y
    # run over loop
    for i in range(n_times):
        X_augmented = generate_augmented_images(X)
        Y_augmented = generate_augmented_labels(Y)
        X = X_augmented
        Y = Y_augmented
    return X_augmented
```

```python
# call the function to check if the image pairs data is augmented as expected
x = generate_augmented_data(X)
```

```python
Y.shape, x.shape
```

```python
#Make train, test split
x_train, x_test, y_train, y_test = train_test_split(x, Y, test_size=0.25, random_state=42)
```

```python
#TO DO: Find the best number of components for X (x or X_augmented)
pca = PCA()
pca.fit(x_train)
```

```python
# Cumulative variance to find the optimum number of components at elbows
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
# Plot variance
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.title('Explained Variance vs Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.savefig("PCA_CumVar_plot.png", format="png", dpi=300, bbox_inches='tight',cmap='gray')
plt.show()
```

### KNN classifier performance evaluation and hypertuning on training datset

```python
# Modify the pipeline to apply both features and labels augmentation
cachedir = mkdtemp()
memory = Memory(cachedir, verbose=0) 

knn_pipeline = Pipeline([
    # function will reshape the train shapes 
    ('scaler', StandardScaler()),  # Step 1: Normalize the data
#     ('UpsampleX', FunctionTransformer(generate_augmented_data)),  # Step 2: Apply the augmentation function
    ('PCA',PCA(n_components=500)),
    ('knn_classifier', KNeighborsClassifier(n_neighbors=1))  # Step 3: Perform the classification
],memory=memory)
```

```python
# print(x_train.shape, y_train.shape)
```

```python
# Fit the pipeline (X is the features and Y is the target)
knn_pipeline.fit(x_train, y_train)
```

```python
# Evaluate the pipeline
y_pred = knn_pipeline.score(x_test, y_test) * 100
print(y_pred)
```

##### Apply gridsearch cv on the KNN

```python
CV_scores = {}
```

```python
parameters = {'knn_classifier__n_neighbors':range(1,8,2),
              'knn_classifier__metric': ['euclidean', 'manhattan']}

knn_model = GridSearchCV(knn_pipeline, parameters, cv=5)
knn_model.fit(x_train, y_train)
print(knn_model.best_params_)
```

```python
CV_scores["KNN"] = cross_val_score(knn_model, x, Y, cv=5, scoring='accuracy')

knn_model = knn_model.best_estimator_
knn_model.score(x_test, y_test.ravel())
```

```python
#save the KNN model
joblib.dump(knn_model, 'knn_model.joblib')
```

### Random Forest classifier performance evaluation and hypertuning on training datset

```python
cachedir = mkdtemp()
memory = Memory(cachedir, verbose=0)
#Pipline for Random Forest
rf_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=900)),
    ('rf', RandomForestClassifier(n_estimators=100)),    
    ])
#parameters for tuning
parameters = {'pca__n_components': [500, 800, 900], 'rf__n_estimators': [10, 50, 100]}
# Initialise and apply the GridSearchCV
rf_model = GridSearchCV(rf_pipeline, parameters, cv=5, n_jobs=-1)
rf_model.fit(x_train, y_train)
print(rf_model.best_params_)
```

### Support Vector Machine classifier performance evaluation and hypertuning on training datset

### Neural Network/MLP classifier performance evaluation and hypertuning on training datset

#### Let's find the AUC score and plot ROC curve for the the best model

```python
eval_scores = model.predict_proba(x_eval)
eval_scores_same_person = eval_scores[:,1]
fpr, tpr, threshold = roc_curve(y_eval, eval_scores_same_person)
roc_auc = auc(fpr, tpr)
```

```python
print('fpr :', fpr)
print('tpr :', tpr) 
print('threshold :', threshold)
print('roc_auc :', roc_auc)
```

```python
# AUC score
auc_eval = roc_auc_score(y_eval, eval_scores_same_person)
auc_eval
```

```python
# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--', lw=2, label='Random Guess')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.legend(loc='lower right')
plt.savefig("ROC-AUC-Curve.png", format="png", dpi=300, bbox_inches='tight')
plt.show()
```
